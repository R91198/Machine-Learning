```python
import pandas as pd
import numpy as np
%pylab inline
```

```python
df = pd.read_csv(r'C:\Users\Rishabh\Documents\IMS Proschool Data Science\Linear Regression\Advertising.csv')
df.head()
```

```python
X = df[['TV','Radio','Newspaper']]
y = df.Sales
```

```python
import statsmodels.api as sm
```

```python
# Fit an OLS(Ordinary Least Square) model with intercept on TV and Radio
X = sm.add_constant(X)
model = sm.OLS(y,X).fit()
```

This constant value is for the intercept that is present in the equation so that the intercept value is calculated and it is being multiplied by this constant value of 1.
