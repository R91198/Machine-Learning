```python
data1 = data.drop(['parks', 'beaches', 'religious_institutions'],1)
data1.head()
```

```python
km = cluster.KMeans(n_clusters=5)
km.fit(data1)
```

```python
data2 = data1.copy()
data2['cluster'] = km.predict(data1)
```

```python
data2.cluster.value_counts()
```

```python
data2.groupby('cluster')[data1.columns].mean().plot.bar(figsize=(12,6))
```

```python
data1sc = preprocessing.scale(data1)
```

```python
ine = []
for i in range(1,100):
    km = cluster.KMeans(n_clusters=i)
    km.fit(data1sc)
    ine.append(km.inertia_)
```

```python
ks = range(1,50)
plt.figure(figsize=(15,7))
plt.plot(ks, ine[:49], '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```
